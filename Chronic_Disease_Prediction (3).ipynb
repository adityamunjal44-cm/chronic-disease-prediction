```python
train_data = pd.read_csv('Training.csv')
test_data = pd.read_csv('Testing.csv')
```

## Data Preprocessing
This step includes separating features and target variables, and preparing the dataset for model training.

```python
train_data = train_data.drop(['Unnamed: 133'],axis=1)
x_test = test_data.drop(["prognosis"], axis = 1)  
y_test = test_data["prognosis"] 
```

```python
# Training split
x_train = train_data.drop(["prognosis"], axis = 1)  
y_train = train_data["prognosis"] 
```

## Feature Scaling

Standardization is applied to ensure all features contribute equally to the model.

```python
# Standardize the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### Logistic Regression


```python
# Initialize the logistic regression model
log_reg = LogisticRegression(max_iter=1000)

# Train the model
log_reg.fit(x_train, y_train)

# Predict the target variable for the test data
y_pred_log_reg = log_reg.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_log_reg)

# Print the accuracy
print(f"Accuracy: {accuracy * 100:.2f}%")
```
